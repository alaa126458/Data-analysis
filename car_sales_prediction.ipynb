# car sales prediction using linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# Try reading the file with a different encoding
dataset=pd.read_csv('car_purchasing.csv', encoding='latin-1')
dataset.head()
```

```python
dataset.drop(['customer name','customer e-mail','net worth'], axis=1,inplace=True, errors='ignore')
```

```python
dataset.head(5)
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split # import train_test_split from model_selection
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

# Assuming the first column (index 0) of x_train and x_test contains the country names
# We exclude this columnf from scaling
x_train_scaled = sc.fit_transform(x_train[:, 1:]) # Select all rows and columns from index 1 onwards
x_test_scaled = sc.transform(x_test[:, 1:])
```

```python
# If you are trying to predict a continuous target variable use Linear Regression
from sklearn.linear_model import LinearRegression
classifier = LinearRegression()
classifier.fit(x_train_scaled,y_train)
```

```python
y_pred=classifier.predict(x_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared (R2)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
